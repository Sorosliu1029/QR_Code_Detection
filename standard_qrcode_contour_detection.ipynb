```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
%matplotlib inline
```

```python
qr1 = cv2.imread('./test_pictures/qr-1.png')
plt.imshow(qr1)
plt.show()
qr1 = cv2.cvtColor(qr1, cv2.COLOR_BGR2RGB)
qr_tmp = np.copy(qr1)
```

```python
qr_gray = cv2.cvtColor(qr1, cv2.COLOR_RGB2GRAY)
```

```python
qr_gray.shape
```

```python
external_contour = cv2.findContours(qr_gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
external_contour
```

```python
cv2.drawContours(qr1, external_contour[0], 
                                           -1, (0, 0, 255))
plt.imshow(qr1)
plt.show()
qr1 = np.copy(qr_tmp)
```

```python
all_contour = cv2.findContours(qr_gray, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
len(all_contour)
```

```python
cv2.drawContours(qr1, all_contour[0], -1, (0, 255, 0))
plt.imshow(qr1)
plt.show()
qr1 = np.copy(qr_tmp)
```

```python
len(all_contour[0])
```

```python
cv2.drawContours(qr1, all_contour[0][0], -1, (0, 255, 0))
plt.imshow(qr1)
plt.show()
qr1 = np.copy(qr_tmp)
```
